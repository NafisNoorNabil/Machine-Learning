```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3]
y = [0,2,4,6]
plt.plot(x,y, label ="2x", color = "red", linewidth= 2, marker='.', markersize=10, markeredgecolor='black')
#fmt = '[color],[marker],[line]'
# plt.plot(x,y, 'r.-')
plt.title("First graph", fontdict = {'fontname':'sans serif','fontsize':20})
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.xticks(x)
plt.yticks([3,4,5,6,7])
plt.legend()
plt.show()
```

#Figure size and plot saving

```python
x = [0,1,2,3]
y = [0,2,4,6]
plt.figure(figsize=(5,2), dpi=300)
plt.plot(x,y, 'r.-', label ="2x")

x2 = np.arange(0, 4.2, 0.2)
plt.plot(x2,x2**2,"b", label="x^2")

plt.title("First graph", fontdict = {'fontname':'sans serif','fontsize':20})
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.xticks(x)
plt.yticks(y)
# plt.savefig('firstgraoh.png',dpi=300)
plt.legend()
plt.show()
```

#Histogram

```python
bins = [40,50,60,70,80,90,100]
plt.hist(fifa.Overall, bins=bins, color="#abcdef")
plt.xlabel("Overall")
plt.ylabel("Player Count")
plt.xticks(bins)
plt.show()
```

#Pie Chart

```python
left = fifa.loc[fifa["Preferred Foot"] == "Left"].count()[0]
right = fifa.loc[fifa["Preferred Foot"] == "Right"].count()[0]
labels = ["Left","Right"]
colors= ['#abcdef','#2d96ad']
explode = (.1,.1)
plt.title("Foot Preference of FIFA 19 Players")
plt.pie([left,right], labels=labels, colors=colors, autopct='%.2f %%', explode=explode)
plt.show()
```

#Box Plot

```python
plt.style.use('default')
barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
labels = ["FC Barcelona","Real Madrid"]
plt.boxplot([barcelona,madrid],labels=labels)
plt.show()
```

#Scatter Plot
